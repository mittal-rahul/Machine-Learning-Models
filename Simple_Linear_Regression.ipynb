## Answer 3

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mat
csvfile = pd.read_csv('Q3Dataset.csv')
fdata = csvfile.filter(items=['PovPct', 'Brth15to17'])
X = fdata.iloc[:, :-1].values
Y = fdata.iloc[:,1].values
# Displaying current observations in graph to business
mat.scatter(X,Y,color = 'blue')
mat.title('female birthrate per thousand vs poverty')
mat.xlabel('Poverty Percent')
mat.ylabel('Birth rate 15 to 17')
mat.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=1/3, random_state = 0)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,Y_train)
# Calculating coeffecient (theta 1)
print('Coefficient = ', reg.coef_)
# Calculating intercept (theta 0)
print('Intercept = ', reg.intercept_)
```

```python
#Visualizing the fitted line to show to business
mat.scatter(X,Y,color = 'blue')
mat.plot(X_train, reg.predict(X_train), color = 'red')
mat.title('female birthrate per thousand vs poverty')
mat.xlabel('Poverty Percent')
mat.ylabel('Birth rate 15 to 17')
mat.show()
```

```python
# calculating error terms and variance score for the model
from sklearn.metrics import mean_squared_error, r2_score
Y_predict = reg.predict(X_test)
error = mean_squared_error(Y_test,Y_predict)
print('Error term (Mean squared error)= %.2f'% error)
# for business terms, their will be variance from the predicted values
print('Variance score = %.2f' % r2_score(Y_test, Y_predict))
```
